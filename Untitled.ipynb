```python
# 1: Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from xgboost import XGBRegressor
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout


print("Libraries loaded successfully.")
```

```python
# 2: Load Dataset

DATA_PATH = "MicrosoftStock.csv"   

df = pd.read_csv(DATA_PATH)
print("Dataset Shape:", df.shape)
df.head()
```

```python
#  3: Clean & Preprocess Data

df.columns = [c.strip() for c in df.columns]
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').set_index('date')

# Interpolate missing values
df = df.interpolate(method='time').ffill().bfill()

df.head()
```

```python
#  4: Feature Engineering

def add_technical_indicators(df):
    data = df.copy()

    # SMA
    data['SMA_20'] = data['close'].rolling(20).mean()
    data['SMA_50'] = data['close'].rolling(50).mean()

    # EMA
    data['EMA_20'] = data['close'].ewm(span=20, adjust=False).mean()

    # Bollinger Bands
    data['BB_mid'] = data['close'].rolling(20).mean()
    data['BB_std'] = data['close'].rolling(20).std()
    data['BB_upper'] = data['BB_mid'] + 2 * data['BB_std']
    data['BB_lower'] = data['BB_mid'] - 2 * data['BB_std']

    # RSI
    delta = data['close'].diff()
    up = delta.clip(lower=0)
    down = -1 * delta.clip(upper=0)
    roll_up = up.rolling(14).mean()
    roll_down = down.rolling(14).mean()
    RS = roll_up / (roll_down + 1e-9)
    data['RSI_14'] = 100 - (100 / (1 + RS))

    return data.fillna(method='ffill').fillna(method='bfill')

df = add_technical_indicators(df)
df.tail()
```

```python
# 5: EDA

plt.figure(figsize=(12,5))
plt.plot(df.index, df['close'])
plt.title("MSFT Close Price")
plt.grid(True)
plt.show()

plt.figure(figsize=(12,5))
plt.plot(df['close'], label='Close')
plt.plot(df['SMA_20'], label='SMA20')
plt.plot(df['SMA_50'], label='SMA50')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12,5))
plt.plot(df['RSI_14'])
plt.title("RSI 14")
plt.grid(True)
plt.show()
```

```python
#  6: Prepare Data

features = [
    'open','high','low','close','volume',
    'SMA_20','SMA_50','EMA_20',
    'BB_upper','BB_lower','RSI_14'
]

model_df = df[features].dropna()

# Train-test split (time-based)
split_idx = int(len(model_df) * 0.8)
train_df = model_df.iloc[:split_idx]
test_df  = model_df.iloc[split_idx:]

print("Train size:", train_df.shape)
print("Test size:", test_df.shape)

# Scaling
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_df)
test_scaled  = scaler.transform(test_df)
```

```python
#  7: Create supervised X/y for ML 

def create_supervised(df_scaled, original_df):
    X = []
    y = []
    close_idx = original_df.columns.get_loc('close')

    for i in range(1, len(df_scaled)):
        X.append(df_scaled[i-1])
        y.append(df_scaled[i][close_idx])

    return np.array(X), np.array(y)

X_train, y_train = create_supervised(train_scaled, train_df)
X_test,  y_test  = create_supervised(test_scaled,  test_df)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
```

```python
# 8: Train Basic ML Models

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Models trained.")
```

```python
# 9: XGBoost 

xgb_available = False

import xgboost as xgb
xg = xgb.XGBRegressor(n_estimators=100)
xg.fit(X_train, y_train)
y_pred_xgb = xg.predict(X_test)
xgb_available = True
print("XGBoost trained.")

```

```python
# 10: Evaluation 

def inverse_close(vals, scaler, df):
    idx = df.columns.get_loc('close')
    dmin = scaler.data_min_[idx]
    dmax = scaler.data_max_[idx]
    return vals * (dmax - dmin) + dmin

y_test_actual     = inverse_close(y_test, scaler, test_df)
y_pred_lr_actual  = inverse_close(y_pred_lr, scaler, test_df)
y_pred_rf_actual  = inverse_close(y_pred_rf, scaler, test_df)

# RMSE 
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

print("Linear Regression:")
print(" MAE :", mean_absolute_error(y_test_actual, y_pred_lr_actual))
print(" RMSE:", rmse(y_test_actual, y_pred_lr_actual))

print("\nRandom Forest:")
print(" MAE :", mean_absolute_error(y_test_actual, y_pred_rf_actual))
print(" RMSE:", rmse(y_test_actual, y_pred_rf_actual))

# XGBoost evaluation 
if xgb_available:
    y_pred_xgb_actual = inverse_close(y_pred_xgb, scaler, test_df)
    print("\nXGBoost:")
    print(" MAE:", mean_absolute_error(y_test_actual, y_pred_xgb_actual))
    print(" RMSE:", rmse(y_test_actual, y_pred_xgb_actual))
```
